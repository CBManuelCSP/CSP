### Exercicio 2: Conversións de Tipo

A partires dos seguintes elementos:

```python
numero_str = "1234"
numero_float = 45.67
lista_numeros = [1, 2, 3, 4, 5]
```

Realiza as seguintes conversións:

1. numero_str a enteiro
2. numero_float a enteiro (qué pasa coa parte decimal?)
3. lista_numeros a tupla e a conxunto
4. Crea un string que combine numero_str e numero_float

```python
# String a entero
numero_entero = int(numero_str)

# Float a entero (los decimales son truncados)
decimal_entero = int(numero_float)

# Lista a tupla y conjunto
tupla = tuple(lista_numeros)
conjunto = set(lista_numeros)

# String combinado
combinado = f"{numero_str} - {numero_float}"

print(f"""STRING A ENTERO: {numero_entero}
FLOAT A ENTERO: {decimal_entero}
TUPLA: {tupla}
CONJUNTO: {conjunto}
STRING COMBINADO: {combinado}""")
```

### Exercicio 3: Cadeas

A partires desta cadea:

```python
texto = "  Python é unha linguaxe de programación super famosa  "
```

Realiza estas operacións:

1. Elimina espazos ao comezo e ao final
2. Convirte a maiúsculas
3. Reempraza "linguaxe" por "lang"
4. Divide o string en palabras
5. Atopa a posición de "programación"

```python
sin_espacios = texto.strip()
print(f"SIN ESPACIOS: {sin_espacios}")

mayus = sin_espacios.upper()
print(f"MAYÚSCULAS: {mayus}")

reemplazar = sin_espacios.replace("linguaxe","lang")
print(f"REEMPLAZO: {reemplazar}")

division_en_palabras = sin_espacios.split(" ")
print(f"TEXTO DIVIDIDO EN PALABRAS: {division_en_palabras}")

posicion = division_en_palabras.index("programación")
print(f"POSICIÓN DE PROGRAMACIÓN: {posicion}")
```

### Exercicio 7: Lista de Diccionarios

Crea unha lista de diccionarios con información de estudantes:

```python
estudantes = [
    {"nome": "Ana", "idade": 20, "nota": 8.5},
    {"nome": "Carlos", "idade": 22, "nota": 7.8},
    {"nome": "María", "idade": 21, "nota": 9.2}
]
```

Realiza estas operacións:

1. Engade un nopvo estudante.
2. Atopa o estudiante con maior nota.
3. Calcula o promedio de idades.
4. Crea unha lista só cos nomes.
5. Ordena os estudantes por nota (descendente).

```python
nuevo_estudiante = {"nome": "Juan", "idade": 25, "nota": 6}
estudantes.append(nuevo_estudiante)

maior_nota=0
maior_nota = max(el["nota"] for el in estudantes)
print(maior_nota)

promedio_edades = sum(el["idade"] for el in estudantes) / len(estudantes)
print(f"EDAD PROMEDIO: {promedio_edades:.2f}")

lista_nomes = [el["nome"] for el in estudantes]
print(lista_nomes)

estudantes.sort(key=lambda x: x["nota"], reverse=True)
print(estudantes)
```

### Exercicio 8: Conxuntos

A partir destes conxuntos:

```python
conx_a = {1, 2, 3, 4, 5}
conx_b = {4, 5, 6, 7, 8}
```

Realiza estas operacións:

1. Unión de ambos conxuntos
2. Intersección
3. Diferencia (A - B)
4. Diferencia simétrica
5. Verifica si 3 está en A

```python
union = conx_a | conx_b
print("UNIÓN:", union)

interseccion = conx_a & conx_b
print("INTERSECCIÓN:", interseccion)

diferencia = conx_a - conx_b
print("A - B: ", diferencia)

dif_simetrica = conx_a ^ conx_b
print("DIFERENCIA SIMÉTRICA: ", dif_simetrica)

print("¿3 en A?: ", 3 in conx_a)
```

### Exercicio 9: Programa Integrador

Crea un programa que:

1. Pida ao usuario o seu nome, idade e cidade de residencia
2. Almacene esa información nun diccionario
3. Convirta a idade a enteiro
4. Cree unha lista de hobbies (pregunta por 3 polo menos)
5. Amose toda a información formateada de xeito lexible

```python
usuario = {
    "nome": input("Introduce o teu nome: "),
    "idade": int(input("Introduce a túa idade: ")),  # Convierto a entero aqui
    "cidade": input("Introduce a túa cidade de residencia: ")
}

hobbies = []
print("Introduce polo menos 3 hobbies:")
for i in range(3):
    hobby = input(f"Hobby {i+1}: ")
    hobbies.append(hobby)

usuario["hobbies"] = hobbies

print("\n--- INFORMACIÓN ---")
print(f"NOMBRE: {usuario['nome']}")
print(f"EDAD: {usuario['idade']}")
print(f"CIUDAD: {usuario['cidade']}")
print("HOBBIES:")
for h in usuario["hobbies"]:
    print(f"- {h}")
```

### Exercicio 10: Combinación de Tipos Complexos

Temos un Sistema de biblioteca:

```python
biblioteca = {
    "libros": [
        {
            "titulo": "Cien años de soledad",
            "autor": "Gabriel García Márquez", 
            "ano": 1967,
            "xeneros": {"Realismo mágico", "Novela"},
            "disponible": True
        },
        {
            "titulo": "1984",
            "autor": "George Orwell",
            "ano": 1949,
            "xeneros": {"Ciencia ficción", "Distopía"},
            "disponible": False
        }
    ],
    "usuarios": {
        "001": {"nome": "Ana", "libros_prestados": ["1984"]},
        "002": {"nome": "Carlos", "libros_prestados": []}
    }
}
```

Realiza estas operacións:

1. Engade un novo libro.
2. Marca un libro como non dispoñible.
3. Atopa todos os libros dun xénero.
4. Libros dispoñibles por autor.
5. Usuario con máis libros prestados.

```python
nuevo_libro = {
    "titulo": "El Principito",
    "autor": "Antoine de Saint-Exupéry",
    "ano": 1943,
    "xeneros": {"Fábula", "Infantil"},
    "disponible": True
}
biblioteca["libros"].append(nuevo_libro)

for libro in biblioteca["libros"]:
    if libro["titulo"] == "El principito":
        libro["disponible"] = False

libros_genero = [el["titulo"] for el in biblioteca["libros"] if "Fábula" in el["xeneros"]]
print(libros_genero)

libros_dispoñibles_autor = [libro["titulo"] for libro in biblioteca["libros"] if libro["autor"] == "Antoine de Saint-Exupéry" and libro["disponible"]]

mas_libros_prestados = max(biblioteca["usuarios"].items(), key=lambda x: len(x[1]["libros_prestados"]))
print(mas_libros_prestados[1]["nome"])
```
