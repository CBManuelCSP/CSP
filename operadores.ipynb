### Exercicio 1: Operadores Aritméticos

Calcula a área e perímetro dun rectángulo:
- Pide ao usuario base e altura.
- Calcula área.
- Calcula perímetro.
- Utiliza `//` e `%`.
- Amosa os resultados utilizando diferentes formatos de cadeas.

```python
base = float(input("Introduce la base: "))
altura = float(input("Introduce la altura: "))

area = base * altura
perimetro = 2 * (base + altura)

# Diferentes formatos de cadena:

# 1. Usando f-strings
print(f"Área do rectángulo: {area}")

# 2. Usando formato con %
print("Perímetro do rectángulo: %d" % perimetro)

# 3. Usando str.format()
print("Área dividido entre 3: {} (cociente {}) e resto {}".format(area, area//3, area%3))
print("Perímetro dividido entre 4: {} (cociente {}) e resto {}".format(perimetro, perimetro//4, perimetro%4))
```

### Exercicio 2: Operadores de Asignación

Simula unha conta bancaria:
- Saldo inicial: 1000
- Usa += para depositar 500
- Usa -= para retirar 300
- Usa *= para aplicar interés do 10%
- Usa /= para dividir entre 2 (metade para aforros)
- Amosa saldo despois de cada operación
- Utiliza diferentes formatos de cadeas.

```python
# Saldo inicial
saldo = 1000

# Depositar
saldo += 500
print("Tras depositar: %d" % saldo)

# Retirar
saldo -= 300
print("Tras retirar: {}".format(saldo))

# Aplicar interés
saldo *= 1.10
print(f"Tras aplicar interés: {saldo:.2f}")

# Dividir
saldo /= 2
print("Tras dividir: ", saldo)
```

### Exercicio 3: Operadores de Comparación

Sistema de verificación de acceso:
- Pide idade e se é socio ou non.
- Verifica se idade >= 18 e <= 65
- Verifica se idade <> 0
- Usa todos estes operadores: >, <, >=, <=, ==, !=
- O mesmo: utiliza diferentes formatos de cadeas para amosar os resultados.

```python
edad = int(input("Introduce tu edad: "))
socio = input("Eres socio? (si/no): ").strip().lower()

if socio != "si":
    print("No eres socio")
else:
    print("Eres socio")

if edad==0:
    print("La edad no puede ser %d" % edad)

print(f"Eres {'menor' if edad<18 else 'mayor'} de edad")

if edad < 18:
    print("Debes ser mayor de edad")

if edad > 65:
    print("Tienes ",edad," años. No puedes ser mayor de 65 años")

if edad >= 18 and edad <= 65:
    print("Tienes {} años, por lo que estás en el rango correcto de edad para entrar".format(edad))
```

### Exercicio 4: Operadores Lóxicos

Determina se un ano é bisesto:
- Divisible entre 4.
- Non divisible entre 100, excepto se é divisible entre 400.
- Non encadenes condicionais senón condicións.
- Exemplo: 2024 (bisesto), 1900 (non bisesto), 2000 (bisesto).

```python
anho = int(input("Di un año: "))

bisiesto = (anho%4==0) and ((anho % 100 != 0) or (anho % 400 == 0))

print(f"{'Bisiesto' if bisiesto else 'No bisiesto'}")
```

### Exercicio 6: Operadores de Pertenza

Buscador de vogais en texto:
- Pide unha frase ao usuario.
- Usa `in` para verificar se contén cada vogal.
- Usa `not in` para verificar se non contén números.
- Conta cantas vogais diferentes ten.

```python
frase = input("Introduce una frase: ").lower()

vocales = ["a","e","i","o","u"]
vocales_encontradas = set()

for v in frase:
    if v in vocales:
        vocales_encontradas.add(v)

for v in vocales_encontradas:
    print(f"La vocal {v} está en la frase")

if all(d not in frase for d in "0123456789"):
    print("A frase non contén números.")
else:
    print("A frase contén algún número.")

print(f"Número de vogais diferentes na frase: {len(vocales_encontradas)}")
```

### Exercicio 8: Walrus

Validador de contrasinal con walrus:
- Pide contrasinal ata que cumpra requisitos.
- Lonxitude >= 8 caracteres.
- Contén maiúscula e número.
- Usa walrus para non repetir len(contrasinal).

```python
while (password := input("Introduce una contraseña: ")):
    longitud = len(password) >= 8
    mayus = any(c.isupper() for c in password)
    numero = any(c.isdigit() for c in password)

    if longitud and mayus and numero:
        print("Contraseña válida!")
        break
    else:
        print("Contraseña NO válida. Debe ter al menos 8 caracteres, una mayúscula y un número.\n")
```

### Exercicio 9: Combinación de Operadores

Calculadora de IMC (xa a fixemos pero desta volta é máis completa):
- peso / altura²
- Clasifica segundo táboa OMS
- Baixo peso: < 18.5
- Normal: 18.5 - 24.9
- Sobrepeso: 25 - 29.9
- Obesidade: >= 30
- Usa operadores de comparación e lóxicos
- Opcional: walrus para evitar repetir cálculo

```python
peso = float(input("Introduce el peso (kg): "))
altura = float(input("Introduce la altura (m): "))

if (imc := peso / altura**2) < 18.5:
    clasificacion = "Bajo peso"
elif 18.5 <= imc <= 24.9:
    clasificacion = "Normal"
elif 25 <= imc < 30:
    clasificacion = "Sobrepeso"
else:
    clasificacion = "Obesidad"

print(f"IMC: {imc:.2f}")
print(f"Clasificación según la OMS: {clasificacion}")
```

### Exercicio 10: Analizador de Texto

Analiza un texto introducido polo usuario:
1. Lonxitude.
2. Se contén certas palabras.
3. Conta vogais.
4. Verifica se é palíndromo.
5. Verifica se texto é `None`.
6. Concatena.

```python
texto = input("Introduce un texto: ")

if texto is None or texto == "":
    print("No se ha introducido ningún texto.")
else:
    print("Longitud del texto:", len(texto))

    palabras_buscar = ["hola", "mundo", "python"]

    for palabra in palabras_buscar:
        if palabra in texto.lower():
            print(f"Contiene la palabra: '{palabra}'")

    vocales = "aeiouáéíóú"
    contador_vocales = sum(1 for letra in texto.lower() if letra in vocales)
    print("Número de vocales:", contador_vocales)

    texto_lista = texto.split(" ")
    texto_sin_espacios = "".join(texto_lista).lower()
    if texto_sin_espacios == texto_sin_espacios[::-1]:
        print("El texto es un palíndromo.")
    else:
        print("El texto no es un palíndromo.")

    texto_concatenado = texto + " - texto analizado"
    print("Texto concatenado:", texto_concatenado)
```

### Xogo de Adiviñanza

Xogo onde:
- Número secreto aleatorio (1-100).
- Aportar Pistas ("é maior", "é menor").
- Contador de intentos.
- Utiliza Walrus.
- Preguntar se se quere xogar outra vez.

```python
import random

numero = random.randint(1, 100)
seguir = "s"

while seguir.lower() == "s":
    while (adivinanza := int(input("Introduce un número: "))) != numero:
        print("Es mayor" if adivinanza < numero else "Es menor")

    seguir = input("Quieres jugar otra vez? (s/n): ").strip()

print("Has acertado")
```

### Validador de Email

Valida email con:
- Que conteña unha `@`.
- Que conteña un `.` despois da `@`.
- Lonxitude >= 5.
- Non conteña espazos en branco.
- Dominio pertenza a unha lista de dominios válida.

```python
def comprobar_email(email,dominios_validos):
    if "@" in email:
        email_separado = email.split("@")
        nombre = email_separado[0]
        dominio = email_separado[1]
    else:
        return False
    
    if "." not in dominio:
        return False
    
    if len(nombre)<5:
        return False
    
    if " " in email:
        return False
    
    if dominio not in dominios_validos:
        return False

    return True

dominios_validos = ["gmail.com", "hotmail.com", "yahoo.com", "outlook.com"]

email = input("Introduce un email: ")

if comprobar_email(email,dominios_validos):
    print("Email válido")
else:
    print("Email NO válido")
```
