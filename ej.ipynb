Aqui tenemos un programa para comprobar si un año es bisiesto o no:

Comenzamos pidiendo el año al usuario:

```python
anho = int(input())
```

Comprobamos el año:

```python
if anho%4==0 and anho%100!=0 or anho%400==0:
    bisiesto = True
else:
    bisiesto = False
```

Imprimimos resultado:

```python
print("Bisiesto" if bisiesto else "No bisiesto")
```
